```python
x = np.linspace(0, 1, 10000)
def func(p):
    return np.ceil(p * (1 - p) / (.01 / 1.959963984540054) ** 2)
plt.plot(x, func(x));
```

```python
def func_1(x):
    arr = []
    for i in x:
        arr += [samplesize_confint_proportion(i, .01)]
    return arr
plt.plot(x, func_1(x));
```

```python
np.array(func_1(x)).max()
```
